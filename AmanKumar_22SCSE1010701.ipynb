Viterbi Algorithm for POS Tagging (Python)

```python
import numpy as np

# Hidden States (Weather Conditions)
states = ['Sunny', 'Rainy']

# Observations (Activities done)
observations = ['Walk', 'Shop', 'Clean']

# Start Probabilities (Initial state distribution)
start_prob = {'Sunny': 0.6, 'Rainy': 0.4}

# Transition Probabilities (State changes)
trans_prob = {
    'Sunny': {'Sunny': 0.7, 'Rainy': 0.3},
    'Rainy': {'Sunny': 0.4, 'Rainy': 0.6}
}

# Emission Probabilities (Likelihood of an observation given a state)
emission_prob = {
    'Sunny': {'Walk': 0.6, 'Shop': 0.3, 'Clean': 0.1},
    'Rainy': {'Walk': 0.1, 'Shop': 0.4, 'Clean': 0.5}
}

# Number of observations and states
num_obs = len(observations)
num_states = len(states)

# Viterbi Table (Probability Matrix)
viterbi = np.zeros((num_states, num_obs))
backpointer = np.zeros((num_states, num_obs), dtype=int)

# Step 1: Initialization
for s, state in enumerate(states):
    viterbi[s, 0] = start_prob[state] * emission_prob[state][observations[0]]

# Step 2: Recursion
for t in range(1, num_obs):
    for s, state in enumerate(states):
        max_prob, max_state = max(
            (viterbi[prev_s, t-1] * trans_prob[prev_state][state], prev_s)
            for prev_s, prev_state in enumerate(states)
        )
        viterbi[s, t] = max_prob * emission_prob[state][observations[t]]
        backpointer[s, t] = max_state

# Step 3: Backtracking
best_path = []
last_state = np.argmax(viterbi[:, -1])
best_path.append(states[last_state])

for t in range(num_obs - 1, 0, -1):
    last_state = backpointer[last_state, t]
    best_path.insert(0, states[last_state])

# Output Results
print("Viterbi Table:")
print(viterbi)

print("\nMost Likely Weather Sequence:")
for word, tag in zip(observations, best_path):
    print(f"{word} -> {tag}")
```

```python
import math

# Sample documents (using a list instead of a set to preserve order)
documents = [
    "The food was good",
    "The service was bad",
    "The food was terrible"
]

# Step 1: Tokenize
def tokenize(doc):
    return doc.lower().split()

tokenized_documents = [tokenize(doc) for doc in documents]

# Step 2: Build vocabulary
vocab = sorted(set(word for doc in tokenized_documents for word in doc))

# Step 3: Calculate TF (Term Frequency)
def term_frequency(word, doc):
    return doc.count(word) / len(doc)

tf_scores = []
for doc in tokenized_documents:
    tf_scores.append({word: term_frequency(word, doc) for word in vocab})

# Step 4: Calculate IDF (Inverse Document Frequency)
def inverse_document_frequency(word):
    return math.log10(len(documents) / sum(1 for doc in tokenized_documents if word in doc))

idf_scores = {word: inverse_document_frequency(word) for word in vocab}

# Output
print("Vocabulary:", vocab)
print("\nTF Scores:")
for i, doc_tf in enumerate(tf_scores):
    print(f"Document {i + 1}:", doc_tf)

print("\nIDF Scores:")
for word, score in idf_scores.items():
    print(f"{word}: {score:.4f}")
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfTransformer, TfidfVectorizer
import pandas as pd

# Sample documents
documents = [
    "The food was good",
    "The service was bad",
    "The food was terrible"
]

# Step 1: CountVectorizer (term frequency)
count_vectorizer = CountVectorizer(lowercase=True)
term_freq_matrix = count_vectorizer.fit_transform(documents)

# View vocabulary
vocab = count_vectorizer.get_feature_names_out()
print("Vocabulary:", vocab)

# Convert to DataFrame for better readability
tf_df = pd.DataFrame(term_freq_matrix.toarray(), columns=vocab)
print("\nTF (Term Frequency) Matrix:")
print(tf_df)

# Step 2: IDF and TF-IDF using TfidfTransformer
tfidf_transformer = TfidfTransformer()
tfidf_matrix = tfidf_transformer.fit_transform(term_freq_matrix)

idf_values = tfidf_transformer.idf_
idf_df = pd.DataFrame([idf_values], columns=vocab)
print("\nIDF Values:")
print(idf_df)

tfidf_df = pd.DataFrame(tfidf_matrix.toarray(), columns=vocab)
print("\nTF-IDF Matrix:")
print(tfidf_df)
```
