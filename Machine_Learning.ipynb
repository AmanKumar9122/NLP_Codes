```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

```python
df = pd.read_csv('homeprices (1).csv')
df
```

Data Preprocessing: Fill NA values with median value of a column

```python
df.bedrooms.median()
```

```python
df.bedrooms = df.bedrooms.fillna(df.bedrooms.median())
df
```

```python
reg = linear_model.LinearRegression()
reg.fit(df.drop('price',axis='columns'),df.price)
```

Find price of home with 3000 sqr ft area, 3 bedrooms, 40 year old

```python
reg.predict([[3000, 3, 40]])
```

Find price of home with 2500 sqr ft area, 4 bedrooms, 5 year old

```python
reg.predict([[2500, 4, 5]])
```

```python
d = pd.read_csv('hiring.csv')
d
```

```python
d.experience = d.experience.fillna("zero")
d
```

```python
!pip install word2number # Install the necessary library
from word2number import w2n # Import the library with the standard alias, but import w2n directly
d.experience = d.experience.apply(w2n.word_to_num) # Now w2n.word_to_num is recognized
d
```

```python
import math
median_test_score = math.floor(d['test_score(out of 10)'].mean())
median_test_score
```

```python
d['test_score(out of 10)'] = d['test_score(out of 10)'].fillna(median_test_score)
d
```

```python
reg = linear_model.LinearRegression()
reg.fit(d[['experience','test_score(out of 10)','interview_score(out of 10)']],d['salary($)'])
```

2 yr experience, 9 test score, 6 interview score

```python
reg.predict([[2,9,6]])
```

12 yr experience, 10 test score, 10 interview score

```python
reg.predict([[12,10,10]])
```
